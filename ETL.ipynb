```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None) # MUESTRA TODAS LAS COLUNAS DE UN PANDAS DATAFRAME
from fastapi import FastAPI
import sklearn
```

```python
df1= pd.read_csv(r'C:movies_dataset.csv', sep=',')
df2= pd.read_csv(r'C:credits.csv',sep=',')
movies_dataset= pd.concat([df1, df2], axis=1)
movies_dataset.head()
```

```python
#Revisar campos con valores nulos
movies_dataset.isna().sum()
```

```python
# Rellenar los valores nulos en los campos 'revenue' y 'budget' con 0
movies_dataset['revenue'].fillna(0, inplace=True)
movies_dataset['budget'].fillna(0, inplace=True)
```

```python
# Eliminar filas con valores nulos en 'release date'
movies_dataset.dropna(subset=['release_date'], inplace=True)
```

```python
#Revisar que no haya valores nulos
movies_dataset.isna().sum()
```

```python
movies_dataset.info()
```

```python
# Eliminar las columnas no deseadas
columnas_no_deseadas = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
movies_dataset.drop(columnas_no_deseadas, axis=1, inplace=True)
```

```python
movies_dataset.head()
```

```python
# Convertir la columna 'release date' a tipo de datos de fecha
movies_dataset['release_date'] = pd.to_datetime(movies_dataset['release_date'], errors='coerce')

# Formatear las fechas en el formato 'AAAA-mm-dd'
movies_dataset['release _date'] = movies_dataset['release_date'].dt.strftime('%Y-%m-%d')
```

```python
# Crear la columna 'release_year' a partir del año de la fecha de estreno
movies_dataset['release_year'] = movies_dataset['release_date'].dt.strftime('%Y')
```

```python
movies_dataset.head()
```

```python
movies_dataset.info()
```

```python
# Convertir las columnas 'revenue' y 'budget' a tipo numérico
movies_dataset['revenue'] = pd.to_numeric(movies_dataset['revenue'], errors='coerce')
movies_dataset['budget'] = pd.to_numeric(movies_dataset['budget'], errors='coerce')

# Crear la columna 'return' con el cálculo del retorno de inversión y tratar los valores de cero
movies_dataset['return'] = np.where(
    (movies_dataset['budget'] == 0) | (movies_dataset['revenue'] == 0),
    0,  # Valor predeterminado para los casos en los que budget o revenue sea cero
    movies_dataset['revenue'] / movies_dataset['budget']
)
```

```python
from fastapi import FastAPI
app = FastAPI()

@app.get('/peliculas_idioma')
def peliculas_idioma(idioma: str):
    cantidad = len(df_movies[df_movies['original_language'] == idioma])
    return f"{cantidad} películas fueron producidas en {idioma}" 
```

```python
peliculas_idioma('en')
```

```python
@app.get('/peliculas_duracion')
def peliculas_duracion(pelicula: str):
    pelicula = df_movies[df_movies['title'] == pelicula]
    if len(pelicula) > 0:
        duracion = pelicula['runtime'].values[0]
        anio = pelicula['release_year'].values[0]
        return f"{pelicula['title'].values[0]}. Duración: {duracion}. Año: {anio}"
    else:
        return f"No se encontró la película '{pelicula}'"


```

```python
peliculas_duracion('Jumanji')
```

```python
@app.get('/franquicia')
def franquicia(franquicia: str):
    peliculas = df_movies[df_movies['belongs_to_collection'] == franquicia]
    cantidad_peliculas = len(peliculas)
    ganancia_total = peliculas['revenue'].sum()
    ganancia_promedio = peliculas['revenue'].mean()
    
    return f"La franquicia '{franquicia}' posee {cantidad_peliculas} película(s), una ganancia total de {ganancia_total} y una ganancia promedio de {ganancia_promedio}"
```

```python
franquicia('Harry Potter')
```

```python
@app.get('/peliculas_pais')
def peliculas_pais(pais: str):
    peliculas = df_movies[df_movies['production_countries'].apply(lambda x: pais in str(x))]
    cantidad_peliculas = len(peliculas)
    
    return f"Se produjeron {cantidad_peliculas} película(s) en el país '{pais}'"

```

```python
peliculas_pais('United States of America')
```

```python
@app.get('/productoras_exitosas')
def productoras_exitosas(productora: str):
    peliculas = df_movies[df_movies['production_companies'].apply(lambda x: isinstance(x, list) and productora in x)]
    cantidad_peliculas = len(peliculas)
    revenue_total = peliculas['revenue'].sum()

    return f"La productora '{productora}' ha tenido un revenue de {revenue_total} y ha realizado {cantidad_peliculas} película(s)"

```

```python
productoras_exitosas('Warner Bros.')
```

```python
@app.get('/get_director')
def get_director(nombre_director):
    movies_list = []
    total_return = 0
    count = 0

    for _, movie in df_movies.iterrows():
        directors = movie['directors']

        if nombre_director in directors:
            title = movie['title']
            release_date = movie['release _date']
            individual_return = movie['return']
            costo = movie['budget']
            ganancia = movie['revenue']

            movie_details = {
                'title': title,
                'release _date': release_date,
                'individual_return': individual_return,
                'costo': costo,
                'ganancia': ganancia
            }

            total_return += individual_return
            count += 1

            movies_list.append(movie_details)

    exito_director = total_return / count if count > 0 else 0
    movies_list.insert(0, {'exito_director': exito_director})

    return movies_list


```

```python
get_director('Joe Johnston')
```
